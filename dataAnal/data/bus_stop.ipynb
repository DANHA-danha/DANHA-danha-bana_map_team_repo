# 서울시 버스정류소 위치정보

## 1. 데이터 불러오기 및 확인

```python
df_raw = pd.read_csv(
    "/Users/danha/Desktop/IT/project/team_project_brew_map/git/data/서울시 버스정류소 위치정보 (1).csv",
    encoding="cp949"
)

print("Shape:", df_raw.shape)
print("Columns:", df_raw.columns.tolist())
df_raw.head()
```

```python
print('=== 데이터 타입 ===')
print(df_raw.dtypes)

print('=== 결측값 현황 ===')
print(df_raw.isnull().sum())

```

## 2. 핵심 데이터 추출

```python
clean_cols = ['정류소번호', '정류소명', 'X좌표', 'Y좌표']
df_raw = df_raw[clean_cols]

print(f'추출 후 Shape: {df_raw.shape}')
df_raw.head()
```

## 3. 열이름 영소문자 구분

```python
col_rename = {
    '정류소번호' : 'bus_stop_id',
    '정류소명'    : 'bus_stop_nm',
    'X좌표'     : 'lng',
    'Y좌표': 'lat'
}
df_raw = df_raw.rename(columns=col_rename)
print('변환된 컬럼명:', df_raw.columns.tolist())
df_raw.head()
```

## 4. type 행 추가

```python
df_raw['type'] = '버스정류장'
df_raw.head()
```

```python
save_path = "/Users/danha/Desktop/IT/project/team_project_brew_map/git/data/bus_stop.csv"

df_raw.to_csv(save_path, index=False, encoding="utf-8-sig")

print("저장 완료:", save_path)
```
