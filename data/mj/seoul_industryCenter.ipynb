```python
import requests 
import os
from dotenv import load_dotenv
import pandas as pd
```

```python
# env 파일 로드
load_dotenv()
```

# 지식산업센터

## 데이터 불러오기

```python
apikey  = os.getenv("MJ_APIKEY")     # 인증키
start_index = "1"    # 요청시작위치
end_index   = "300"  # 요청종료위치

url = f"http://openapi.seoul.go.kr:8088/{apikey}/json/InfoBizcenterApply/{start_index}/{end_index}/"

response = requests.get(url)
print(response.content.decode('utf-8'))
```

```python
data = response.json()
rows = data['InfoBizcenterApply']['row']
print(rows)

center_df = pd.DataFrame(rows)
center_df
```

## 데이터 전처리

```python
# 필요한 데이터만 추출하고, 열이름 모두 소문자로 통일
center_df = center_df[['IDX', 'FAC_NM', "ADDRESS"]]
center_df.columns = center_df.columns.str.lower()

center_df
```

```python
# idx: float → int 로 변경
center_df['idx'] = center_df['idx'].astype(int)

# 값 중복 여부 확인
print(center_df['idx'].duplicated().sum()) # 0
```

```python
# 공장 이름과 주소가 모두 같은 행이 있는지 확인
print(center_df.duplicated(subset=['fac_nm', 'address']).sum()) # 1

dup_all_mask = center_df.duplicated(subset=['fac_nm', 'address'], keep=False)
center_df[dup_all_mask].sort_values(['fac_nm', 'address'])
```

```python
# 공장 이름과 주소 중복 행 제거
center_df = center_df.drop_duplicates(subset=['fac_nm', 'address'], keep='first').reset_index(drop=True)
center_df
```

```python
# 주소 이상치 제거
center_df["address"] = (
    center_df["address"]
    .str.replace(r"(,|\n|외).*$", "", regex=True)  # 첫 , 또는 \n 또는 외 이후 전부 삭제
    .str.strip()
)

center_df
```

```python
print(center_df['address'].str.contains(',').sum()) # 외 = 0, \n = 0, , = 1
print(center_df[center_df['address'].str.contains(',', na=False)]) # idx = 202

cond = center_df['idx'] == 202
center_df.loc[cond, 'address'] = '성수동2가 269-25' # 첫 주소로 대체

center_df.loc[cond, :]
```

```python
# type 변수 추가
center_df['type'] = '지식산업센터'
center_df
```

### 주소 → 위도, 경도

```python
kakao_key  = os.getenv("KAKAO_REST_APIKEY")     # 인증키
```

```python
# test
url = "	https://dapi.kakao.com/v2/local/search/address.json" #요청할 url 주소
headers = {"Authorization": f"KakaoAK {kakao_key}"} #REST API 키(유효한 키)
params = {"query": "성수동2가 269-25"}  # 입력할 주소
resp = requests.get(url,headers=headers, params=params).json() #카카오 API 요청
                      
print(resp) #API 결과 출력
resp['documents'][0]['road_address']
```

```python
url = "	https://dapi.kakao.com/v2/local/search/address.json" #요청할 url 주소
headers = {"Authorization": f"KakaoAK {kakao_key}"} #REST API 키(유효한 키)

for idx, add in center_df['address'].items():
    params = {"query": add}
    resp = requests.get(url, headers=headers, params=params).json()

    docs = resp.get('documents', [])
    if not docs:
        continue

    road = docs[0].get('road_address')  # 없을 수도 있음

    if road:
        center_df.loc[idx, 'road_address'] = road.get('address_name')
        center_df.loc[idx, 'lat'] = road.get('y') # Y 좌표값, 경위도인 경우 위도(latitude)
        center_df.loc[idx, 'lng'] = road.get('x') # X 좌표값, 경위도인 경우 경도(longitude)
    else:
        # road_address가 없으면 지번주소(address) 기준으로라도 좌표 저장
        addr = docs[0].get('address')
        if addr:
            center_df.loc[idx, 'road_address'] = addr.get('address_name')
            center_df.loc[idx, 'lat'] = addr.get('y')
            center_df.loc[idx, 'lng'] = addr.get('x')

center_df
```

```python
print(center_df['road_address'].str.contains('서울').sum())

not_seoul = center_df[~center_df['road_address'].str.contains('서울')]
not_seoul
```

```python
# idx 재설정 및 정렬
center_df.sort_values(by = 'idx', inplace=True)
center_df['idx'] = range(1, len(center_df) + 1) # idx 번호 재부여
center_df.reset_index(drop=True, inplace=True) # index 재설정
center_df
```
